```python
# Predictive stats compare to Official results
# Differences and accuracy

# Official and predicted results
official_results = {
    "Donald J. Trump": 57.94,
    "Kamala Harris": 42.06
}

predicted_results = {
    "Donald J. Trump": 57.39,
    "Kamala Harris": 42.61
}

# Calculate differences
differences = {}
for candidate in official_results:
    differences[candidate] = abs(predicted_results[candidate] - official_results[candidate])

# Print results
for candidate in official_results:
    official = official_results[candidate]
    predicted = predicted_results[candidate]
    difference = differences[candidate]
    
    if predicted > official:
        sign = "+"
    else:
        sign = "-"
    
    print(f"{candidate}:")
    print(f"  Official Percentage: {official:.2f}%")
    print(f"  Predicted Percentage: {predicted:.2f}%")
    print(f"  Difference = {sign}{difference:.2f}%")
    print()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Candidates
candidates = ['Donald J. Trump', 'Kamala Harris']

# Official Results
official_results = [57.94, 42.06]

# Predicted Results
predicted_results = [57.39, 42.61]

# Create a bar chart
x = np.arange(len(candidates))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, official_results, width, label='Official Results')
bars2 = ax.bar(x + width/2, predicted_results, width, label='Predicted Results')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage (%)')
ax.set_title('Comparison of Official Results and Predicted Results')
ax.set_xticks(x)
ax.set_xticklabels(candidates)
ax.legend()

# Add value labels on top of the bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Candidates
candidates = ['Donald J. Trump', 'Kamala Harris']

# Official Results
official_results = [57.94, 42.06]

# Predicted Results
predicted_results = [57.39, 42.61]

# Calculate Differences
differences = [predicted - official for predicted, official in zip(predicted_results, official_results)]

# Create a scatter plot
plt.figure(figsize=(8, 6))

# Plot Official Results
plt.scatter(candidates, official_results, color='g', label='Official Results', s=100, marker='o')
# Plot Predicted Results
plt.scatter(candidates, predicted_results, color='b', label='Predicted Results', s=100, marker='x')

# Add lines to show differences
for i in range(len(candidates)):
    plt.plot([candidates[i], candidates[i]], [official_results[i], predicted_results[i]], color='gray', linestyle='--')

# Add labels and title
plt.title('Comparison of Official and Predicted Results')
plt.xlabel('Candidates')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)  # Set y-axis limit for better visualization
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add a horizontal line at y=0
plt.grid()

# Add value labels on the points
for i, (official, predicted, diff) in enumerate(zip(official_results, predicted_results, differences)):
    plt.text(i, official + 1, f'Official: {official:.2f}%', ha='center', color='g')
    plt.text(i, predicted + 1, f'Predicted: {predicted:.2f}%', ha='center', color='b')
    plt.text(i, (official + predicted) / 2, f'Diff: {diff:.2f}%', ha='center', color='gray')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
